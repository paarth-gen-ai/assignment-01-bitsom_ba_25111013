```python
# Import libraries
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')
print('Model loaded successfully!')
```

```python
# Define 10 sentences across 3 domains
# Cricket (1-3), Cooking (4-6), Cybersecurity (7-10)
sentences = [
    "The bowler took three wickets in one over",
    "The batsman hit a six to win the match",
    "The cricket team won the championship trophy",
    "She kneaded the dough for fresh bread",
    "Add salt and pepper to taste while cooking",
    "The recipe requires olive oil and garlic",
    "The firewall blocked unauthorized network access",
    "Hackers used phishing to steal credentials",
    "Two-factor authentication prevents account breaches",
    "Encryption protects sensitive data from attackers"
]

print(f"Total sentences: {len(sentences)}")
for i, s in enumerate(sentences):
    print(f"{i+1}: {s}")
```

```python
# Generate embeddings for all sentences
embeddings = model.encode(sentences)

print(f"Embedding shape: {embeddings.shape}")
print(f"Each sentence is represented by a {embeddings.shape[1]}-dimensional vector")
```

```python
# Compute 10x10 cosine similarity matrix
similarity_matrix = cosine_similarity(embeddings)

print("Cosine Similarity Matrix (10x10):")
print(np.round(similarity_matrix, 2))
```

```python
# Plot heatmap of cosine similarity matrix
plt.figure(figsize=(12, 10))

# Create short labels for display
short_labels = [
    "Cricket1", "Cricket2", "Cricket3",
    "Cooking1", "Cooking2", "Cooking3",
    "Sec1", "Sec2", "Sec3", "Sec4"
]

sns.heatmap(similarity_matrix, annot=True, fmt='.2f', cmap='YlOrRd',
            xticklabels=short_labels, yticklabels=short_labels,
            square=True, linewidths=0.5)
plt.title('Cosine Similarity Heatmap - 10 Sentences')
plt.xlabel('Sentence')
plt.ylabel('Sentence')
plt.tight_layout()
plt.show()

print("Heatmap shows intra-domain similarity (diagonal blocks) vs inter-domain (off-diagonal)")
```

```python
# Semantic search query
query = "The bowler took three wickets in one over"

# Encode the query
query_embedding = model.encode([query])

# Compute similarity with all sentences
query_similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get indices sorted by similarity (descending)
sorted_indices = np.argsort(query_similarities)[::-1]

print(f"Query: '{query}'\n")
print("Top 2 semantic matches:\n")

for i in sorted_indices[:2]:
    print(f"Rank {sorted_indices.tolist().index(i)+1}:")
    print(f"  Sentence: '{sentences[i]}'")
    print(f"  Similarity Score: {query_similarities[i]:.4f}\n")
```

## Summary

This demo showed how sentence embeddings capture semantic meaning:
- Sentences in the same domain (e.g., Cricket) have higher similarity
- The cosine similarity heatmap visualizes semantic clustering
- Querying with a sentence retrieves semantically similar results
